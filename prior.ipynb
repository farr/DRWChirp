```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
prng = np.random.default_rng(42)

n_samples_init = 10000000

n_grid = 5
w_min = 0.0
dw = 1

w0 = prng.normal(w_min, dw, n_samples_init)
w0 = w0[w0 >= 0]

rand_int = prng.integers(low=0, high=n_grid, size=len(w0))
w = w0 + rand_int * dw

ax = plt.gca()
sns.histplot(w, stat='density', element='step', alpha=0.3, ax=ax)
ax.set_xlabel(r'$\left(\omega_0 - \omega_{\rm min}\right)/\Delta\omega_0$')
ax.set_ylabel(r'$p(\omega_0 \mid \Delta\omega_0)$')
ax.set_xticks(np.arange(0, n_grid + 3, 1), labels=np.arange(0, n_grid + 3, 1))
ax.set_xlim(0, n_grid + 2.5)
ax.grid(axis='x', linestyle=':', linewidth=2)

# Add vertical line and label
ax.axvline(x=n_grid, color='red', linestyle='--')
ax.text(n_grid + 0.1, ax.get_ylim()[1]*0.9, r'$N_{\omega_0}$', color='red', va='top')

plt.savefig("omega-prior.pdf", bbox_inches='tight')
```
